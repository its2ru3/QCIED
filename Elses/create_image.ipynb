## Image creation

```python
import numpy as np
from PIL import Image

# 1. Define the RGB data as a 3D NumPy array
# Example: a 2x2 image with Red (top-left), Green (top-right), Black (bottom-left and bottom-right)
rgb_matrix = np.array([
    [[1,2,1], [2,3,2], [2,6,0], [0,7,2]],
    [[3,5,0], [3,4,1], [2,4,2], [3,7,5]],
    [[4,7,4], [2,6,1], [7,7,7], [6,6,6]],
    [[3,7,5], [3,5,0], [0,7,2], [0,2,0]]
], dtype=np.uint8) # Data type must be uint8 for image compatibility

 
# 2. Convert the NumPy array to a Pillow Image object
img = Image.fromarray(rgb_matrix, 'RGB')

# 3. Save the image to a file
img.save("input.jpg")

# Optional: Display the image
img.show()
```

```python
import numpy as np
from PIL import Image

# 1. Create your 3-bit RGB matrix (values 0-7)
# Example: 100x100 random image
rgb_3bit = np.random.randint(0, 8, (10, 10, 3), dtype=np.uint8)

# 2. Rescale to 8-bit (0-255) for display
# This ensures 0 remains black and 7 becomes full intensity (255)
rgb_8bit = (rgb_3bit * (255 / 7)).astype(np.uint8)

# 3. Create and show the image
img = Image.fromarray(rgb_8bit, 'RGB')
img.show()
```
